```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import logit

# Load the Excel file
file_path = '/Users/katherinekrieger/Desktop/Wald_Test_Trial_treatment.xlsx'
data = pd.read_excel(file_path)

# Check the data structure
print(data.head())

# Ensure the dependent variable is binary (0 or 1)
data['recommendation_presence'] = data['recommendation_presence'].astype(int)

# Set reference category
data['race_ethnicity'] = pd.Categorical(data['race_ethnicity'], categories=['black', 'white','hispanic', 'asian'], ordered=False)

# Fit the logistic regression model using the formula API
model = logit('recommendation_presence ~ race_ethnicity', data=data).fit()

# Print the summary of the model
print(model.summary())

# Perform Wald test for significance of each independent variable
wald_test_results = model.wald_test_terms(scalar=True)
print(wald_test_results)
```
